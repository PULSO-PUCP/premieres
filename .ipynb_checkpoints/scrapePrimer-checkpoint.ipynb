```python
# coding: utf-8
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "https://es.wikipedia.org/wiki/Anexo:Presidentes_del_Consejo_de_Ministros_del_Per%C3%BA"

try:
    response = requests.get(url)
    response.raise_for_status()
    response.encoding = 'utf-8'
    soup = BeautifulSoup(response.text, 'html.parser')

    all_tables = soup.find_all('table', {'class': 'wikitable'})
    dfs = []  # List to store individual DataFrames

    for table in all_tables:
        data = []
        headers = [th.text.strip() for th in table.find_all('th')]
        # Append headers (even if empty)
        data.append(headers)

        rows = table.find_all('tr')[1:]
        for row in rows:
            cols = [td.text.strip() for td in row.find_all('td')]
            if cols:
                data.append(cols)

        # Create DataFrame, allowing for mismatched column counts
        if data:
            df = pd.DataFrame(data[1:], columns=data[0] if data[0] else None)
            dfs.append(df)

    if dfs:
        final_df = pd.concat(dfs, ignore_index=True, sort=False)
        print(final_df.head())
    else:
        print("No tables with class 'wikitable' found on the page.")

except requests.exceptions.RequestException as e:
    print(f"Error fetching the URL: {e}")
except Exception as e:
    print(f"An error occurred: {e}")
```

```python

def split_at_july28(df, start_col='start', end_col='end', id_col='id'):
    """
    Splits each spell at every July 28 between start and end dates using Andersen–Gill.
    Returns a DataFrame of intervals with columns:
      - original columns
      - interval_start, interval_end
    """
    intervals = []
    for _, row in df.iterrows():
        start = row[start_col]
        end = row[end_col] if pd.notnull(row[end_col]) else pd.Timestamp('2025-05-08')
        # Build list of July 28 cut-points after the start date
        years = range(start.year + 1, end.year + 1)
        cuts = [pd.Timestamp(year=y, month=7, day=28) for y in years if pd.Timestamp(year=y, month=7, day=28) < end]
        current_start = start
        for cut in cuts:
            new_row = row.copy()
            new_row['interval_start'] = current_start
            new_row['interval_end'] = cut
            intervals.append(new_row)
            current_start = cut
        # final interval
        new_row = row.copy()
        new_row['interval_start'] = current_start
        new_row['interval_end'] = end
        intervals.append(new_row)
    return pd.DataFrame(intervals)

# Usage example:
intervals = split_at_july28(primesmodel)
```

```python
intervals
```

```python
# Compute Fujimori regime dummy for each interval based on interval_start
intervals['regime_fujimori'] = intervals['interval_start'].between(
    pd.Timestamp('1992-04-05'), pd.Timestamp('2000-07-28')
).astype(int)

# Compute speech window indicator (±45 days around July 28)
def in_speech_window(date):
    speech_day = pd.Timestamp(year=date.year, month=7, day=28)
    return int(abs((date - speech_day).days) <= 45)

intervals['speech_window'] = intervals['interval_start'].apply(in_speech_window)
```

```python

# Create event indicator: 1 if interval_end equals original spell end, 0 otherwise
AG_intervals['event'] = ((AG_intervals['interval_end'] == AG_intervals['end']) & AG_intervals['end'].notna()).astype(int)

#  Compute duration in days
AG_intervals['interval_duration'] = (AG_intervals['interval_end'] - AG_intervals['interval_start']).dt.days
```
